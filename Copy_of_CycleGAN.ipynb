# Training

-   `python train.py --dataroot ./datasets/horse2zebra --name horse2zebra --model cycle_gan`

Change the `--dataroot` and `--name` to your own dataset's path and model's name. Use `--gpu_ids 0,1,..` to train on multiple GPUs and `--batch_size` to change the batch size. I've found that a batch size of 16 fits onto 4 V100s and can finish training an epoch in ~90s.

Once your model has trained, copy over the last checkpoint to a format that the testing model can automatically detect:

Use `cp ./checkpoints/horse2zebra/latest_net_G_A.pth ./checkpoints/horse2zebra/latest_net_G.pth` if you want to transform images from class A to class B and `cp ./checkpoints/horse2zebra/latest_net_G_B.pth ./checkpoints/horse2zebra/latest_net_G.pth` if you want to transform images from class B to class A.

```python
from PIL import Image as PILImage
import glob
import os

# Paths
image_dir = 'results/style_vangogh_pretrained/test_latest/images'
fake_images = glob.glob(os.path.join(image_dir, '*_fake.png'))

# Output directory
gif_output_dir = 'gifs'
os.makedirs(gif_output_dir, exist_ok=True)

for fake_path in fake_images:
    real_path = fake_path.replace('_fake', '_real')

    # Open and resize
    fake_img = PILImage.open(fake_path).resize((256, 256))
    real_img = PILImage.open(real_path).resize((256, 256))

    # Create GIF frames
    frames = [real_img, fake_img]

    # File name for GIF
    gif_name = os.path.basename(fake_path).replace('_fake.png', '.gif')
    gif_path = os.path.join(gif_output_dir, gif_name)

    # Save as GIF
    frames[0].save(
        gif_path,
        save_all=True,
        append_images=frames[1:],
        duration=800,  # milliseconds per frame
        loop=0
    )

    print(f"GIF saved: {gif_path}")
```

```python
# 1) Install required libs (run once)
!pip install scikit-image matplotlib pandas

# 2) Python code: compute SSIM and PSNR and plot
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.io import imread
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd

# Paths - change if needed
orig_dir = 'originals'
monet_dir = 'monet_outputs'
vangogh_dir = 'vangogh_outputs'

# Helper to find matching pairs
def basename_no_ext(path):
    return os.path.splitext(os.path.basename(path))[0]

# collect original files
orig_files = sorted(glob.glob(os.path.join(orig_dir, '*')))

rows = []
for orig_path in orig_files:
    name = basename_no_ext(orig_path)
    # expected output names (adjust suffix if different)
    monet_path = os.path.join(monet_dir, name + '_fake_B.png')
    vangogh_path = os.path.join(vangogh_dir, name + '_fake_B.png')
    # if outputs have same base name without suffix, try alternative
    if not os.path.exists(monet_path):
        alt = os.path.join(monet_dir, name + '.png')
        if os.path.exists(alt): monet_path = alt
    if not os.path.exists(vangogh_path):
        alt = os.path.join(vangogh_dir, name + '.png')
        if os.path.exists(alt): vangogh_path = alt

    # read and convert to grayscale for SSIM (or keep RGB by passing multichannel=True)
    orig = imread(orig_path)
    # some images might be grayscale; convert shapes for SSIM consistency
    # compute for RGB if possible
    def safe_read(path):
        if not os.path.exists(path):
            return None
        img = imread(path)
        return img

    monet = safe_read(monet_path)
    vangogh = safe_read(vangogh_path)

    for style_name, out in [('Monet', monet), ('VanGogh', vangogh)]:
        if out is None:
            print(f"Skipping {name} for {style_name}: output not found.")
            continue
        # If shapes differ, resize to smallest common shape (simple crop)
        h = min(orig.shape[0], out.shape[0])
        w = min(orig.shape[1], out.shape[1])
        orig_cropped = orig[:h, :w]
        out_cropped = out[:h, :w]
        # If images are color, compute SSIM with multichannel=True
        multich = True if orig_cropped.ndim == 3 and out_cropped.ndim == 3 else False
        try:
            s = ssim(orig_cropped, out_cropped, multichannel=multich)
        except TypeError:
            s = ssim(orig_cropped, out_cropped)  # fallback
        p = psnr(orig_cropped, out_cropped, data_range=out_cropped.max() - out_cropped.min() if out_cropped.max()!=out_cropped.min() else 255)
        rows.append({'image': name, 'style': style_name, 'ssim': float(s), 'psnr': float(p)})

# Aggregate results
df = pd.DataFrame(rows)
print('Computed for', len(df), 'pairs')
summary = df.groupby('style').agg({'ssim':'mean','psnr':'mean'}).reset_index()
print(summary)

# Save CSV
df.to_csv('ssim_psnr_per_image.csv', index=False)
summary.to_csv('ssim_psnr_summary.csv', index=False)

# Plot grouped bar chart
styles = summary['style'].tolist()
ssim_vals = summary['ssim'].tolist()
psnr_vals = summary['psnr'].tolist()

x = np.arange(len(styles))
width = 0.35
fig, ax = plt.subplots(figsize=(7,4))
rects1 = ax.bar(x - width/2, ssim_vals, width, label='SSIM')
rects2 = ax.bar(x + width/2, psnr_vals, width, label='PSNR')

ax.set_ylabel('Value')
ax.set_title('Average SSIM and PSNR by Style')
ax.set_xticks(x)
ax.set_xticklabels(styles)
ax.legend()
# annotate bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x()+rect.get_width()/2, height),
                    xytext=(0,3), textcoords='offset points', ha='center', va='bottom', fontsize=9)
autolabel(rects1); autolabel(rects2)

plt.tight_layout()
plt.savefig('ssim_psnr_summary.png', dpi=200)
plt.show()

print('Saved ssim_psnr_per_image.csv, ssim_psnr_summary.csv and ssim_psnr_summary.png')
```

# Testing

-   `python test.py --dataroot datasets/horse2zebra/testA --name horse2zebra_pretrained --model test --no_dropout`

Change the `--dataroot` and `--name` to be consistent with your trained model's configuration.

> from https://github.com/junyanz/pytorch-CycleGAN-and-pix2pix:
> The option --model test is used for generating results of CycleGAN only for one side. This option will automatically set --dataset_mode single, which only loads the images from one set. On the contrary, using --model cycle_gan requires loading and generating results in both directions, which is sometimes unnecessary. The results will be saved at ./results/. Use --results_dir {directory_path_to_save_result} to specify the results directory.

> For your own experiments, you might want to specify --netG, --norm, --no_dropout to match the generator architecture of the trained model.
